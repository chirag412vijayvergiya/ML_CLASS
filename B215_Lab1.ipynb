```python
# What is the use of Grouping ? 
# Grouping helps in understanding the data better.
# It helps in finding the relationship between different variables.
# df_group_two = df_group_two.to_frame()
df_gptest = df[['drive-wheels','body-style','price']]
grouped_test1 = df_gptest.groupby(['drive-wheels','body-style'],as_index=False).mean()
grouped_test1
```

```python
grouped_pivot = grouped_test1.pivot(index='drive-wheels',columns='body-style')  
grouped_pivot
```
