```python
import pandas as pd
import numpy as np
import seaborn as sns

df = pd.read_excel("EDA Cars.xlsx")
```

```python
# 2. Find the size of data frame
print("Size of data frame is: ", df.size)
```

```python
# 3. Performed head operation to see first 5 observations
df.head()
```

```python
# 4. Find number of missing values
print("Missing Data : ")
df.isnull().sum()



```

```python
df1 = df.isnull()
for column in df1.columns.values.tolist():
    print(column)
    print(df1[column].value_counts())
    print("")
```

```python
df["INCOME"] = df["INCOME"].fillna(df["INCOME"].median())  
df["TRAVEL TIME"] = df["TRAVEL TIME"].fillna(df["TRAVEL TIME"].median())
df["CAR AGE"] = df["CAR AGE"].fillna(df["CAR AGE"].median())
df["MILES CLOCKED"] = df["MILES CLOCKED"].fillna(df["MILES CLOCKED"].median())
```

```python
# Categorical Data
ms = df["MARITAL STATUS"].value_counts().idxmax()
df["MARITAL STATUS"] = df["MARITAL STATUS"].replace(np.nan, ms)
```

```python
df["SEX"] = df["SEX"].fillna(df["SEX"].mode()[0])
```

```python
ms = df["EDUCATION"].value_counts().idxmax()
df["EDUCATION"] = df["EDUCATION"].replace(np.nan, ms)
```

```python
df["JOB"] = df["JOB"].fillna(df["JOB"].mode()[0])
```

```python
df["USE"] = df["USE"].fillna(df["USE"].mode()[0])
```

```python
df["CAR TYPE"] = df["CAR TYPE"].fillna(df["CAR TYPE"].mode()[0])
```

```python
df["CITY"] = df["CITY"].fillna(df["CITY"].mode()[0])
```

```python
# 7. Remove duplicate records
df2  = df.drop_duplicates()
df2
```

```python
df2.reset_index(drop=True, inplace=True)
```

```python
df2
```

```python
# 8. Draw boxplot for INCOME and find out insights
sns.boxplot(data = df2, y = "INCOME")
```

```python
# Remove outlier using IQR(Inter-Quartile Range) for INCOME, check by using boxplot
Q3, Q1 = np.percentile(df2["INCOME"], [75 ,25])
print(Q3, Q1)
```

```python
IQR = Q3 - Q1
print(IQR)

upper_boundary = Q3 + 1.5 * IQR
lower_boundary = Q1 - 1.5 * IQR

print(upper_boundary, lower_boundary)
```

```python
df2[(df2["INCOME"] > upper_boundary)].count()
```

```python
new_df = df2[df2["INCOME"] < upper_boundary]
new_df
```

```python
sns.boxplot(data = new_df, y = "INCOME")    
```

```python
new_df["MARITAL STATUS"] = new_df["MARITAL STATUS"].astype("category")
new_df["CAR TYPE"] = new_df["CAR TYPE"].astype("category")
new_df["CITY"] = new_df["CITY"].astype("category")
new_df["EDUCATION"] = new_df["EDUCATION"].astype("category")
new_df["JOB"] = new_df["JOB"].astype("category")
new_df["USE"] = new_df["USE"].astype("category")
new_df["SEX"] = new_df["SEX"].astype("category")
```
